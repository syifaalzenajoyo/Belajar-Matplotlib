## Membuat Sample Dataset

- Memanggil `np.linspace()` yang akan menghasilkan sederet bilangan yang disertai 3 parameter.
    - `np.linspace(-3.0, 3.0, 100)` akan menghasilkan sederet bilangan dengan nilai terkecil -3.0, nilai terbesar 3.0, dan banyak point dataya adalah 100.
- Memanggil `np.meshgrid(n, n)` yang akan menghasilkan data surface atau area dan dua buah array data dua dimensi yang akan ditampung oleh variabel x dan y.
- Membuat variabel z untuk manampung `np.sqrt(x**2 + y**2)`

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## 3D Wireframe

- Membuat object figure dengan memanggil fungsi `figure()` yang akan menghasilkan suatu object figure yang akan kita tampung di variabel `fig`
- Membuat object axes dengan memanggil fungsi `axes()` yang disertai oleh parameter optional `projection='3d` untuk menghasilkan 3D axes dan akan ditampung oleh variabel `ax`.
- Memanggil `ax.plot_wireframe(x, y, z, color='orange')` untuk menjadikan nilai dari variabel x, y, dan z sebagai nilai koordinat atau nilai yang akan direpresentasikan pada 3D Wireframe dengan color orange.
- Mengatur judul, xlabel, ylabel, dan zlabel.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

## 3D Surface Plot

- Membuat object figure dengan memanggil fungsi `figure()` yang akan menghasilkan suatu object figure yang akan kita tampung di variabel `fig`
- Membuat object axes dengan memanggil fungsi `axes()` yang disertai oleh parameter optional `projection='3d` untuk menghasilkan 3D axes dan akan ditampung oleh variabel `ax`.
- Memanggil `ax.plot_surface(x, y, z, cmap='viridis')` untuk menjadikan nilai dari variabel x, y, dan z sebagai nilai koordinat atau nilai yang akan direpresentasikan pada 3D Surface Plot dengan cmap viridis.
- Mengatur judul, xlabel, ylabel, dan zlabel.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
