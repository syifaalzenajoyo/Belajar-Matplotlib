## 3D Axes

- Membuat object figure dengan memanggil fungsi `figure()` yang akan menghasilkan suatu object figure yang akan kita tampung di variabel `fig`
- Membuat object axes dengan memanggil fungsi `axes()` yang disertai oleh parameter optional `projection='3d` untuk menghasilkan 3D axes dan akan ditampung oleh variabel `ax`.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

## 3D Line Plot

- Menyiapkan 3 buah variabel yang akan di plotting
    - `z = np.linspace(0, 1, 100)` akan menghasilkan deret bilangan dengan nilai terkecil 0, nilai terbesar 1, dan banyak data pointnya adalah 100.
    - `x = z * np.sin(20 * z)` akan menampung hasil perkalian dari z dengan np.sin(20*z)
    - `y = z * np.cos(20 * z)` akan menampung hasil perkalian dari z dengan np.cos(20*z)

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

- Membuat object figure dengan memanggil fungsi `figure()` yang akan menghasilkan suatu object figure yang akan kita tampung di variabel `fig`
- Membuat object axes dengan memanggil fungsi `axes()` yang disertai oleh parameter optional `projection='3d` untuk menghasilkan 3D axes dan akan ditampung oleh variabel `ax`.
- Memanggil `ax.plot(x, y, z, color='red')` untuk menjadikan nilai dari variabel x, y, dan z sebagai nilai koordinat atau nilai yang akan direpresentasikan pada plotting dengan color red.
- Mengatur judul, xlabel, ylabel, dan zlabel dengan memanggil `ax.set(title='Contoh 3D Line Plot', xlabel='Sumbu X', ylabel='Sumbu Y', zlabel='Sumbu Z')`

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='red')

ax.set(title='Contoh 3D Line Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```

## 3D Scatter Plot

- Menyiapkan 3 buah variabel yang akan di plotting
    - `z = np.linspace(0, 1, 100)` akan menghasilkan deret bilangan dengan nilai terkecil 0, nilai terbesar 1, dan banyak data pointnya adalah 100.
    - `x = z * np.sin(20 * z)` akan menampung hasil perkalian dari z dengan np.sin(20*z)
    - `y = z * np.cos(20 * z)` akan menampung hasil perkalian dari z dengan np.cos(20*z)

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

- Membuat object figure dengan memanggil fungsi `figure()` yang akan menghasilkan suatu object figure yang akan kita tampung di variabel `fig`
- Membuat object axes dengan memanggil fungsi `axes()` yang disertai oleh parameter optional `projection='3d` untuk menghasilkan 3D axes dan akan ditampung oleh variabel `ax`.
- Memanggil `ax.scatter(x, y, z, color='green')` untuk menjadikan nilai dari variabel x, y, dan z sebagai nilai koordinat atau nilai yang akan direpresentasikan pada 3D scatter plotting dengan color green.
- Mengatur judul, xlabel, ylabel, dan zlabel dengan memanggil `ax.set(title='Contoh 3D Line Plot', xlabel='Sumbu X', ylabel='Sumbu Y', zlabel='Sumbu Z')`

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='green')

ax.set(title='Contoh 3D Scatter Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```
