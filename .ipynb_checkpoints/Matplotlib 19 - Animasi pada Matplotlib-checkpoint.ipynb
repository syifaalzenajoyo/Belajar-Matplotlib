## Animasi pada Matplotlib

- membuat object figure dengan memanggil `plt.figure()` yang ditampung oleh variabel 'fig'
- menentukan limit batas atas dan bawah sumbu x dan sumbu y dengan memanggil `plt.xlim((0,4))` (sumbu x memiliki jangkauan 0 sampai 4) dan `plt.ylim((-2,2))`(sumbu y memiliki jangkauan -2 sampai 2).
- memanggil `plt.plot([])[0]` yang disertai dengan parameter berupa sekumpulan data yang akan diplotting. Pemanggilan index [0] untuk membantu kita dalam mengendalikan garis pada object matplotlib.lines.Line2D.
- membuat fungsi `animate()` yang disertai oleh parameter `i`. Fungsi tersebut akan menyiapkan nilai x dan y.
    - x = np.linspace(0, 4, 1000) artinya akan membentuk deret bilangan dengan nilai terkecilnya 0, nilai terbesarnya 4, dan banyak bilangannya adalah 1000 data point.
    - y = np.sin(2 * np.pi * (x-0.01*i))
    - line.set_data(x,y) untuk memanggil garisnya
- `FuncAnimation(fig, animate, frames=200, interval=20)`
    - FuncAnimation() merupakan fungsi utama untuk menghasilkan animasi.
    - Parameter pertama yaitu figure `fig`
    - Parameter kedua yaitu function`animate`
    - Parameter ketiga menentukan jumlah frame yang akan kita gunakan pada animasi, semakin tinggi nilai framenya semakin halus animasi yang dihasilkan.
    - Parameter empat menentukan intervalnya, semakin besar nilai intervalnya semakin lambat animasi yang dihasilkan.

```python
fig = plt.figure()
plt.xlim((0,4))
plt.ylim((-2,2))

line = plt.plot([])[0]

def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x-0.01*i))
    line.set_data(x,y)
    return line,

anim = FuncAnimation(fig, animate, frames=200, interval=20)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)

plt.close()
```
