## Membuat Sample Dataset

Membuat data set yang berisikan dictionary yang terdiri dari pasangan keys yang akan menjadi nama kolom (nama, usia, gender, provinsi, n_anak, dan n_pets) beserta valuesnya.

```python
df = pd.DataFrame({
    'nama': ['cecep', 'karti', 'bejo', 'tejo', 'asep', 'wati', 'amat'],
    'usia': [23, 78, 22, 19, 45, 33, 20],
    'gender': ['L', 'P', 'L', 'L', 'L', 'P', 'L'],
    'provinsi':
    ['jabar', 'jatim', 'jabar', 'jatim', 'jabar', 'jateng', 'jateng'],
    'n_anak': [2, 0, 0, 3, 2, 1, 4],
    'n_pets': [5, 1, 0, 5, 2, 2, 3]
})

df
```

## Scatter Plot

- Membentuk scatter plot dari data set yang telah dibuat.

    - Melakukan visualisasi data dari data frame dengan memanggil method `plot()` yang diikuti oleh 4 parameter.
        - Parameter pertama `kind=='scatter'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu scatter plot.
        - Parameter kedua `x='n_anak'` untuk menentukan kolom acuan untuk sumbu-x
        - Parameter ketiga `y='n_pets'` untuk menentukan kolom acuan untuk sumbu-y
        - Parameter keempat `color='red'` untuk menentukan warna dari marker yang berwarna merah.
    > df.plot(kind='scatter', x='n_anak', y='n_pets', color='red')

    - Menentukan title pada scatter plot dengan judulnya yaitu  `Scatter Plot`
    > plt.title('Scatter Plot')
    
    - Menentukan label pada sumbu-x dengan labelnya yaitu `Jumlah Anak`
    > plt.xlabel('Jumlah Anak')

    - Menentukan label pada sumbu-y dengan labelnya yaitu `Jumlah Pets`
    > plt.ylabel('Jumlah Pets')
        
    - Menampilkan hasil plotting
    > plt.show()

```python
df.plot(kind='scatter', x='n_anak', y='n_pets', color='red')

plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Pets')
plt.show()
```

## Histogram

- Membentuk jostogram dari data set yang telah dibuat.

    - Melakukan visualisasi data dari kolom `'usia'` pada data frame dengan melakukan filtering `[['usia']]` kemudian memanggil method `plot()` yang diikuti oleh 2 parameter.
        - Parameter pertama `kind=='hist'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu histogram.
        - Parameter kedua `
        bins=[0, 20, 40, 60, 80, 100]` 
    > df[['usia']].plot(kind='hist', bins=[0, 20, 40, 60, 80, 100])

    - Menentukan title pada scatter plot dengan judulnya yaitu  `Histogram`
    > plt.title('Histogram')
    
    - Menentukan label pada sumbu-x dengan labelnya yaitu `Usia`
    > plt.xlabel('Usia')

    - Menentukan label pada sumbu-y dengan labelnya yaitu `Jumlah`
    > plt.ylabel('Jumlah')
        
    - Menampilkan hasil plotting
    > plt.show()

```python
df[['usia']].plot(kind='hist', bins=[0, 20, 40, 60, 80, 100])

plt.title('Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```

### Menamplikan Multiple Line Plot pada Satu Axes yang Sama

- Memanggil `plt.gca()` untuk mengembalikan object axes (gca= get current axes) yang ditampung oleh variabel `ax`

- Melakukan visualisasi data pertama dari data frame dengan memanggil method `plot()` yang diikuti oleh 5 parameter.
    - Parameter pertama `kind=='line'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu line plot.
    - Parameter kedua `x='nama'` untuk menentukan kolom acuan untuk sumbu-x
    - Parameter ketiga `y='n_anak'` untuk menentukan kolom acuan untuk sumbu-y
    - Parameter keempat `color='green'` untuk menentukan warna dari marker yang berwarna hijau.
    - Parameter kelima `ax=ax` untuk menentukan axes yang digunakan dalam menampilkan hasil plotting.
    
    > df.plot(kind='line', x='nama', y='n_anak', color='green', ax=ax)
    
- Melakukan visualisasi data pertama dari data frame dengan memanggil method `plot()` yang diikuti oleh 4 parameter.
    - Parameter pertama `kind=='line'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu line plot.
    - Parameter kedua `x='nama'` untuk menentukan kolom acuan untuk sumbu-x
    - Parameter ketiga `y='n_pets'` untuk menentukan kolom acuan untuk sumbu-y
    - Parameter keempat `color='red'` untuk menentukan warna dari marker yang berwarna hijau.
    - Parameter kelima `ax=ax` untuk menentukan axes yang digunakan dalam menampilkan hasil plotting.
        
    > df.plot(kind='line', x='nama', y='n_pets', color='red', ax=ax)

- Menentukan title pada scatter plot dengan judulnya yaitu  `Line Plot`
  > plt.title('Line Plot')

- Menentukan label pada sumbu-y dengan labelnya yaitu `Jumlah`
  > plt.ylabel('Jumlah')
        
- Menampilkan hasil plotting
  > plt.show()

```python
ax = plt.gca()

df.plot(kind='line', x='nama', y='n_anak', color='green', ax=ax)
df.plot(kind='line', x='nama', y='n_pets', color='red', ax=ax)

plt.title('Line Plot')
plt.ylabel('Jumlah')
plt.show()
```

## Bar Plot

- Membentuk bar plot dari data set yang telah dibuat.

    - Melakukan visualisasi data dari data frame dengan memanggil method `plot()` yang diikuti oleh 3 parameter.
        - Parameter pertama `kind=='bar'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu bar plot.
        - Parameter kedua `x='nama'` untuk menentukan kolom acuan untuk sumbu-x
        - Parameter ketiga `y='usia'` untuk menentukan kolom acuan untuk sumbu-y
    > df.plot(kind='bar', x='nama', y='usia')

    - Menentukan title pada bar plot dengan judulnya yaitu  `Bar Plot`

```python
df.plot(kind='bar', x='nama', y='usia')

plt.title('Bar Plot')
plt.ylabel('Usia')
plt.show()
```

- Mengelompokkan data frame berdasarkan *provinsi* yang pada bagian kolom *nama* nya akan dikenakan sebuah fungsi agregasi yaitu `count()` untuk menghitung jumlah barisnya dan melakukan visualisasi data dari data frame dengan memanggil method `plot()` yang diikuti oleh 1 parameter.
    - Parameter pertama `kind=='bar'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu bar plot.
> df.groupby('provinsi')['nama'].count().plot(kind='bar')

- Menentukan title pada bar plot dengan judulnya yaitu  `Bar Plot`
  > plt.title('Bar Plot')

- Menentukan label pada sumbu-y dengan labelnya yaitu `Jumlah`
  > plt.ylabel('Jumlah')
        
- Menampilkan hasil plotting
  > plt.show()

```python
df.groupby('provinsi')['nama'].count().plot(kind='bar')

plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

## Stacked bar plot dengan dua level group by

- Mengelompokkan data frame berdasarkan *provinsi* dan *gender* dan melakukan visualisasi data dari data frame dengan memanggil method `plot()` yang diikuti oleh 2 parameter.
    - Parameter pertama `kind=='bar'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu bar plot.
    - Parameter kedua `stacked=True`
> df.groupby(['provinsi', 'gender']).size().unstack().plot(kind='bar', stacked=True)

- Menentukan title pada bar plot dengan judulnya yaitu  `Bar Plot`
  > plt.title('Bar Plot')

- Menentukan label pada sumbu-y dengan labelnya yaitu `Jumlah`
  > plt.ylabel('Jumlah')
        
- Menampilkan hasil plotting
  > plt.show()

```python
df.groupby(['provinsi', 'gender']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

- Mengelompokkan data frame berdasarkan *gender* dan *provinsi* dan melakukan visualisasi data dari data frame dengan memanggil method `plot()` yang diikuti oleh 2 parameter.
    - Parameter pertama `kind=='bar'` untuk menentukan jenis plotting yang ingin ditampilkan yaitu bar plot.
    - Parameter kedua `stacked=True`
> df.groupby(['gender', 'plot']).size().unstack().plot(kind='bar', stacked=True)

- Menentukan title pada stacked bar plot dengan judulnya yaitu  `Stacked Bar Plot`
  > plt.title('Stacked Bar Plot')

- Menentukan label pada sumbu-y dengan labelnya yaitu `Jumlah`
  > plt.ylabel('Jumlah')
        
- Menampilkan hasil plotting
  > plt.show()

```python
df.groupby(['gender', 'provinsi']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Stacked Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```
