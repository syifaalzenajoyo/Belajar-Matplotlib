```python
x = np.linspace (1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## Menerapkan Temporary Styling

- Melakukan plotting dengan menggunakan `ggplot` style
- Memanggil `plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')` yang akan menghasilkan plotting nilai dari `np.sin(np.linspace(0, 2 * np.pi))` berwarna merah dengan bentuk bulatan dan disertai garis.

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

- Menerapkan `dark_background` secara temporary dapat dilakukan dengan memanggil klausa `with`. Plotting yang berada pada scope `with` akan dikenakan `dark_background` yang hanya berlaku pada bagian ini saja.

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')

plt.show()
```

- Melakukan plotting ketiga yang mengacu pada style awal yang tidak dikenakan oleh `dark_background` style untuk membuktikan bahwa plotting sebelumnya itu bersifat temporary dan tidak mengubah plotting awalnya.

```python
plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```
