#Membuat plotting sederhana

Untuk membuat plotting sederhana, kita dapat memanggi method plt.plot yang disertakan dengan deret bilangan, sebagai contoh [2, 5, 7, 11]

Selain itu, untuk menset label y nya kita dapat memanggil plt.ylabel dan diberikan nilai string sebagai lebelnya.

Pemanggiln plt.show() berfungsi untuk memberikan tampilan yang lebih rapih yang tidak disertai keterangan tulisan tambahan yang tidak dibutuhkan yang dihasilkan dari plt.ylabel('sumbu y').

```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()
```

-Pemanggilan fungsi plot() dapat dilakukan dengan menyertakan hanya sebuah deret bilangan berupa list/array ataupun dua buah deret bilangan berupa list/array.

-Bila pemanggilan hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan tersebut akan dijadikan data point untuk nilai pada sumbu y sedangkan nilai pada sumbu x akan secara otomatis dibuatkan sebuah deret bilangan terurut dengan nilai dimulai dari nol.

-Dalam memanggil plt.plot(), kita menyertakan dua paramater berupa deret bilangan. Parameter pertama berkorelasi dengan sumbu x dan parameter kedua berkorelasi dengan sumbu y.

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show()
```

#Pengaturan Format pada plot

-Selain menyertakan dua buah deret bilangan sebagai parameter untuk sumbu x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur warna dan jenis plotting. 

-Pada contoh di bawah ini, kita akan menggunakan parameter ketiga yang bernilai 'ro'. r berarti red yang mengindikasikan warna merah dan markernya berupa karakter o yang mengindikasikan hasilnya berupa titik-titik.

-r = red, b = blue, g = greeen.

-marker karakter o (hasil berupa titik-titik), karakter - (hasil berupa garis), dan karakter * (hasil berupa bintang).

-contoh penggunaan :
g-, ro, b*.

-Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB.

-Nilai default untuk parameter ketiga ini adalah 'b-' (blue dash). 

-Pemanggilan fungsi axis() menyertakan sebuah list [xmin, xmax, ymin, ymax].

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

#Multi plot dalam satu pemanggilan fungsi plot()

-Kita akan mencoba memanggil fungsi np.arange().

Pemanggilan fungsi np.arange() akan membutuhkan 3 buah parameter, yaitu nilai minimum, pembatas nilai maksimum, dan interval nilai.

```python
t = np.arange(0., 5., 0.2)
t
```

-Melakukan pemanggilan plt.plot() sebanyak 3 kali untuk menghasilkan 3 plot, yaitu plot linear, plot kuadrat, plot, dan plot cubic.

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

-Kita juga dapat melakukan multi plot dalam satu pemanggilan fungsi plot(), artinya kita tidak perlu melakukan pemaanggilan plot sebanyak 3 kali untuk menghasilkan 3 plot.

```python
plt.plot(t, t, 'r--', 
         t, t**2, 'bs', 
         t, t**3, 'g^')
plt.show()
```

#Plotting dengan keywords

Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument (kwargs) seperti pada dictionary dan Python Pandas Data Frame.

-Membentuk sebuah dictionary sederhana dengan 3 key yaitu 'a', 'c', dan 'd' dengan value nya akan berkorelasi secara berurutan dengan np.arange() yang jumlahnya 50, np.random.randint() dengan jangkauan 0-50 sebanyak 50 buah, dan np.random.randn() dengan jumlah angkanya ada 50.

-Kemudian, dictionary ini akan ditambah dengan key baru yaitu key 'b' dan 'd'.

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

-Method scatter plot dengan parameter x nya bernilai key 'a', parameter y nya berniali key 'b', dan parameter terakhirnya data=data.

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

-Menampilkan dimensi ke-3 dan ke-4 dengan menggunakan scatter plot dengan representasi warna. Pada contoh di bawah ini, representasi warna akan bergantung pada 'c' dan representasi ukura akan bergantung pada 'd'.

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

#Plotting untuk tipe data categorical

Matplotlib juga mendukung plotting untuk data dengan tipe kategori atau categorical data atau categorical variables.

Berikut ini, kita memiliki dua buah kumpulan data yaitu names yang nilainya adalah ['group_a', 'group_b', 'group_c'] dan values yang nilainya adalah [1, 10, 100].

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

-Plotting dengan menggunakan plt.plot dengan sumbu x nya mengacu pada variabel names dan sumbu y nya mengacu pada variabel values.

```python
plt.plot(names, values)
plt.show()
```

-Plotting dengan menggunakan plt.scatter dengan sumbu x nya mengacu pada variabel names dan sumbu y nya mengacu pada variabel values.

```python
plt.scatter(names, values)
plt.show()
```

-Plotting dengan menggunakan plt.bar dengan sumbu x nya mengacu pada variabel names dan sumbu y nya mengacu pada variabel values.

```python
plt.bar(names, values)
plt.show()
```
