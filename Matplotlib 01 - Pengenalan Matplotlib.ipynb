#Membuat plotting sederhana

-Membuat sebuah figure yang memiliki sebuah axes; melakukan plotting data pada axes

-Membuat sample data

-Sample data yang dibuat terdiri dari x (sekumpulan nilai 1, 2, 3, 4) dan y (sekumpulan nilai 1, 4, 2, 3). Nilai x dan y akan divisualisasikan.

-Membuat sebuah figure dan sebuah axes dengan menggunakan method suplots()

-Method suplots() secara default akan membentuk sebuah objek figure dan sebuah objek axes sehingga kita perlu membuat dua buah variabel (fig dan ax) untuk menampung figure dan axes tersebut.

-Melakukan Plotting/Visualisasi data pada axes dengan memanggil method plot()

-Menggunakan method plot() pada pyplot untuk plotting data.

#Komponen dari Sebuah Matplotlib Figure 



-Figure merupakan media/container dasar untuk melakukan visualisasi data dengan menggunakan matplotlib.

-Dalam satu figure, kita dapat memiliki beberapa axes. Sebuah figure, dapat memiliki lebih dari satu axes.

-Axes merupakan area gambar (area untuk melakukan plotting) pada anatomy of a figure di atas. Axes berada di dalam figure.

-Axis mengindikasikan sumbu (x, y, z). Di dalam sebuah figure, bisa terdiri dari 2 atau 3 axis yaitu x-axis, y-axis, dan z-axis tergantung pada visualisasi data yang akan kita bangun (plotting 2 dimensi atau 3 dimensi).

#Membuat figure dan axes

-Membuat figure tanpa axes

Method figure() dapat membuat hanya satu objek figure tanpa disertai axes. Sehingga, kita hanya perlu membuat satu variabel yaitu fig untuk menampung figure yang dibuat.

Note : untuk membuat sebuah visualisasi data, kita perlu membuat axes.

```python
fig = plt.figure()
```

-Membuat figure yang disertai dengan sebuah axes

Method subplots() akan membentuk sebuah objek figure dan sebuah objek axes. Oleh karena itu, kita perlu membuat dua buah variabel (fig dan ax) untuk menampung figure dan axes tersebut.

```python
fig, ax = plt.subplots()
```

-Membuat figure dengan beberapa axes

Method subplots() yang disertai dengan parameter yaitu plt.suplots(2, 3) akan menghasilkan sebuah figure dengan 2 x 3 = 6 grid axes. 

parameter (2, 3) artinya terdapat 2 buah baris dan 3 buah kolom.

Object figure akan disimpan dalam variabel fig dan 6 object axes akan disimpan dalam variabel axs.

```python
fig, axs = plt.subplots(2, 3)
```

#Melakukan Visualisasi Data dengan Menggunakan Pendekatan OO Style

Ciri dari OO Style adalah secara eksplisit mebuat figures beserta axes dan memanggil methods dari keduanya.

-Menyiapkan data setnya.

Method np.linspace(0, 2, 100) digunakan untuk membuat sejumlah deret bilangan yang ditentukan oleh 3 parameter. 

Parameter pertama  mengindikasikan starting pointnya yaitu 0, parameter kedua merupakan end pointnya yaitu 2,
dan parameter ketiga menunjukkan banyaknya data point yang dibentuk yaitu 100.

```python
x = np.linspace(0, 2, 100)
x
```

-Membuat sebuah figure dan sebuah axes secara eksplisit dengan menggunakan method plt.subplots() dan akan ditampung ke dalam dua buah variabel yaitu fig dan ax.

-Plotting tiga variant data pada axes dengan meggunakan method ax.plot() yang disertai tiga parameter yaitu nilai x, nilai y, dan labelnya.

-method ax.set_xlabel('x label') digunakan untuk menyertakan x-label pada axes dengan label sesuai dengan parameter yang dimasukkan yaitu x label.

-method ax.set_ylabel('y label') digunakan untuk menyertakan y-label pada axes dengan label sesuai dengan parameter yang dimasukkan yaitu y label.

-method ax.set_title('Simple Plot') digunakan untuk menyertakan title pada axes dengan title sesuai dengan parameter yang dimasukkan yaitu Simple Plot.

-method ax.legend() digunakan untuk menyertakan legend.

```python
fig, ax = plt.subplots()

ax.plot(x, x, label = 'linear')
ax.plot(x, x**2, label = 'quadratic')
ax.plot(x, x**3, label = 'cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```

#Melakukan Visualisasi Data dengan Menggunakan Pendekatan pyplot Style

Menggunakan pyplot untuk membuat dan mengelola figures dan axes serta menggunakan fungsi pada pyplot untuk melakukan plotting.  Kita tidak perlu membentuk figure dan axes secara eksplisit.

-Menyiapkan data setnya.

Method np.linspace(0, 2, 100) digunakan untuk membuat sejumlah deret bilangan yang ditentukan oleh 3 parameter. 

Parameter pertama  mengindikasikan starting pointnya yaitu 0, parameter kedua merupakan end pointnya yaitu 2,
dan parameter ketiga menunjukkan banyaknya data point yang dibentuk yaitu 100.

```python
x = np.linspace(0, 2, 100)
x
```

-Plotting tiga variant data pada axes dengan meggunakan method plt.plot() yang disertai tiga parameter yaitu nilai x, nilai y, dan labelnya.

-method plt.xlabel('x label') digunakan untuk menyertakan x-label pada axes dengan label sesuai dengan parameter yang dimasukkan yaitu x label.

-method plt.ylabel('y label') digunakan untuk menyertakan y-label pada axes dengan label sesuai dengan parameter yang dimasukkan yaitu y label.

-method plt.title('Simple Plot') digunakan untuk menyertakan title pada axes dengan title sesuai dengan parameter yang dimasukkan yaitu Simple Plot.

-method plt.legend() digunakan untuk menyertakan legend.

```python
plt.plot(x, x, label = 'linear')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x**3, label = 'cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
```
